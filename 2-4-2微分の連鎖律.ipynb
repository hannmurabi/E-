```python
import numpy as np
```

```python
#シグモイド関数を定義
def sigmoid(a):
  return 1/(1 + np.exp(-a))
```

```python
#二乗和誤差関数を定義
def sse(y,t):
  return 0.5 * (np.sum(y-t)**2)
```

```python
w = np.array([[0.1]])
x = np.array([[3]])
t = np.array([[1]])
```

```python
a = w.dot(x)
a
```

```python
y = sigmoid(a)
y
```

```python
#二乗和誤差を誤差関数Eとして代入
E = sse(y,t)
E
```

```python
#wの更新式を定義する❶
def diff(y,a,x):
  return(y-1)*(1-sigmoid(a))*sigmoid(a)*x

d = diff(y,a,x)
d
```

```python
#wの更新式を定義する❷
w = w - diff(y,a,x)
w
```

```python
#更新されたwの値を使ってもう一度誤差を計算する
a = w.dot(x)
y = sigmoid(a)

#Eは誤差がどれだけ減ったかの確認のために見る
#wの更新式には関わらない
E = sse(y,t)
```

```python
for _ in range(100):
  w = w - diff(y,a,x)
  a = w.dot(x)
  y = sigmoid(a)
  E = sse(y,t)
```
